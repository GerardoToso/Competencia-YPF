```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas_profiling import ProfileReport
import warnings
import plotly.express as px
import chardet
%matplotlib inline
```

```python
# Detección automática de la codificación de los caracteres
with open("/content/drive/MyDrive/DataScience/TRABAJOS PARTICULARES/YPF/DATASET/Dataset participantes (corregido).csv", 'rb') as f:
    enc = chardet.detect(f.read())  # or readline if the file is large

# Lectura del csv
df = pd.read_csv("/content/drive/MyDrive/DataScience/TRABAJOS PARTICULARES/YPF/DATASET/Dataset participantes (corregido).csv",encoding = enc['encoding'], index_col=0)

df.head(5)
```

```python
# Diccionario con todos los valores de delta_WHP correspondientes a cada pozo (incluyendo delta_WHP = 0)
delta_WHP_dic_0 = {}

for i in range(len(df.HIJO.unique())):
  hijo_i = df.HIJO.unique()[i]
  df_hijo_i = df[df.HIJO == hijo_i]
  delta_WHP_dic_0[hijo_i] = df_hijo_i.delta_WHP.tolist()
```

```python
# Diccionario con todos los valores de delta_WHP correspondientes a cada pozo (excluyendo delta_WHP = 0)
delta_WHP_dic = {}
df_aux= df[df.delta_WHP != 0]

for i in range(len(df_aux.HIJO.unique())):
  hijo_i = df_aux.HIJO.unique()[i]
  df_hijo_i = df_aux[df_aux.HIJO == hijo_i]
  delta_WHP_dic[hijo_i] = df_hijo_i.delta_WHP.tolist()
```

```python
for i in delta_WHP_dic:
  plt.figure(figsize=(8, 3))
  sns.kdeplot(data= delta_WHP_dic, x=delta_WHP_dic.get(i),cut = 0,palette='Set1').set(title=i)
```
