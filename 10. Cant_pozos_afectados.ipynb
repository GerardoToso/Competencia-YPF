```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas_profiling import ProfileReport
import warnings
import plotly.express as px
import chardet
%matplotlib inline
```

```python
# Detección automática de la codificación de los caracteres
with open("/content/drive/MyDrive/DataScience/TRABAJOS PARTICULARES/YPF/DATASET/Dataset participantes (corregido).csv", 'rb') as f:
    enc = chardet.detect(f.read())  # or readline if the file is large

# Lectura del csv
df = pd.read_csv("/content/drive/MyDrive/DataScience/TRABAJOS PARTICULARES/YPF/DATASET/Dataset participantes (corregido).csv",encoding = enc['encoding'], index_col=0)

df.sample(5)
```

**- POZOS_AFECT: Cantidad de pozos padre distintos que afectó el pozo hijo en todas sus fracturas**

```python
# Imputacion de valores de cantidad de pozos afectados
for i in range(len(df.HIJO.unique())):
  hijo_i = df.HIJO.unique()[i]
  df_hijo_i = df[df.HIJO == hijo_i]
  padre_unique = len(df_hijo_i.PADRE.unique())
  df.loc[df_hijo_i.index,"POZOS_AFECT"] = padre_unique
```

**- POZOS_POR_FRACTURA: Promedio de pozos padre afectados por el pozo hijo por cada fractura**

```python
# Imputacion de valores de promedio de pozos afectados por fractura
for i in range(len(df.HIJO.unique())):
  hijo_i = df.HIJO.unique()[i]
  df_hijo_i = df[df.HIJO == hijo_i]
  cant_etapas = len(df_hijo_i.ETAPA_HIJO.unique())
  cant_de_afectaciones = df_hijo_i.shape[0]
  prom_pozos_afect_por_fractura = cant_de_afectaciones/cant_etapas
  df.loc[df_hijo_i.index,"POZOS_POR_FRACTURA"] = prom_pozos_afect_por_fractura
```
